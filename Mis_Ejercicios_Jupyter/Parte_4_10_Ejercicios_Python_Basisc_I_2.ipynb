## Ejercicio 1

1. Crea dos variables numericas: un `int` y un `float`
2. Comprueba sus tipos
3. Sumalas en otra nueva
4. ¿De qué tipo es la nueva variable?
5. Elimina las dos primeras variables creadas

```python
# 1. Crear un int (Entero)
edad = 35

# 2. Crear un Float (Decimal)
precio_portatil = 999.99

# 3. Comprueba sus tipos
print(type(edad))  # <class 'int'>
print(type(precio_portatil))  # <class 'float'>

# Correção: usar isinstance() em vez de type() == tipo
if isinstance(edad, int):
    print("Es un entero")

if isinstance(precio_portatil, float):
    print("Es un decimal")

# 4. Sumarlas en una nueva variable
total = edad + precio_portatil
print(f"Total: {total}")

# 5. Tipo de la nueva variable
print(f"Tipo del resultado: {type(total)}")

# 6. Eliminar las dos primeras variables
del edad, precio_portatil

# Verificar que 'total' sigue existiendo
print(f"Total sigue existiendo: {total}")
print("Variables 'edad' y 'precio_portatil' eliminadas")
```

## Ejercicio 2
Escribe un programa para pasar de grados a radianes. Hay que usar `input`. Recuerda que la conversión se realiza mediante

radianes = grados*(pi/180)

```python
# Programa para convertir grados a radianes
print("\n=== CONVERSOR DE GRADOS A RADIANES ===")

try:
    # Solicitar los grados al usuario
    grados = float(input("Introduce los grados: "))
    
    # Fórmula: radianes = grados * (π / 180)
    import math
    radianes = grados * (math.pi / 180)
    
    # Mostrar el resultado
    print(f"\n{grados}° grados equivalen a {radianes} radianes")
    print(f"Valor aproximado: {radianes:.4f} radianes")
except ValueError:
    print("Error: Por favor, introduce un número válido")
```

## Ejercicio 3
Escribe un programa que calcule el area de un paralelogramo (base x altura). También con `input`

```python
# Programa para calcular el área de un paralelogramo
print("\n=== CALCULADORA DE ÁREA DE PARALELOGRAMO ===")
print("Fórmula: Área = base × altura\n")

try:
    # Solicitar datos al usuario
    base = float(input("Introduce la base: "))
    altura = float(input("Introduce la altura: "))
    
    # Calcular el área
    area = base * altura
    
    # Mostrar resultado
    print(f"\nEl área del paralelogramo es: {area}")
except ValueError:
    print("Error: Por favor, introduce números válidos")
```

## Ejercicio 4
Escribe un programa que calcule el area de un paralelogramo (base x altura). También con `input`. En este caso debe ser capaz de admitir valores reales, con decimales, para base y altura en español. (PISTA: Recuerda que tendrás que hacer uso del método replace)

```python
# Programa simple con soporte para formato español
print("\n=== ÁREA DEL PARALELOGRAMO (FORMATO ESPAÑOL) ===")
print("Usa punto (.) o coma (,) para decimales\n")

try:
    # Pedir base y reemplazar coma por punto
    base_texto = input("Base: ")
    base = float(base_texto.replace(',', '.'))
    
    # Pedir altura y reemplazar coma por punto
    altura_texto = input("Altura: ")
    altura = float(altura_texto.replace(',', '.'))
    
    # Calcular
    area = base * altura
    
    # Mostrar resultado en formato español
    area_texto = str(area).replace('.', ',')
    print(f"\n✅ Área = {base_texto} × {altura_texto} = {area_texto}")
    
except ValueError:
    print("Error: Por favor, introduce números válidos (usa punto o coma)")
```
