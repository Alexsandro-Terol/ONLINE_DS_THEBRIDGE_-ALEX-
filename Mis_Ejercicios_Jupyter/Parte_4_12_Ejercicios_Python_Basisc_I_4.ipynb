## Ejercicio 1
Para este ejercicio vamos a poner en práctica [las funciones built in](https://docs.python.org/3/library/functions.html).

1. Calcula el máximo de la lista: [4, 6, 8, -1]
2. Suma todos los elementos de la lista anterior
3. Redondea este float a 3 dígitos decimales: 63.451256965
4. Valor absoluto de: -74

```python
# 1. Calcula el máximo de la lista: [4, 6, 8, -1]
lista = [4, 6, 8, -1]
maximo = max(lista)
print("1. Máximo:", maximo)

# 2. Suma todos los elementos de la lista anterior
suma = sum(lista)
print("2. Suma:", suma)

# 3. Redondea este float a 3 dígitos decimales: 63.451256965
numero = 63.451256965
redondeado = round(numero, 3)
print("3. Redondeado:", redondeado)

# 4. Valor absoluto de: -74
absoluto = abs(-74)
print("4. Valor absoluto:", absoluto)
print()
```

    "En un lugar de la Mancha, de cuyo nombre no quiero acordarme."

2. Pásalo todo a minúsculas
3. Solo la primera letra de cada palabra en mayúscula, el resto en minúscula
4. Crea una lista dividiéndolo por sus espacios
5. Sustituye las comas `,` por puntos y comas `;`

```python
# 1. Pásalo todo a mayúsculas
print("1. Mayúsculas:", texto.upper())

# 2. Pásalo todo a minúsculas
print("2. Minúsculas:", texto.lower())

# 3. Solo la primera letra de cada palabra en mayúscula
print("3. Title:", texto.title())

# 4. Crea una lista dividiéndolo por sus espacios
lista_palabras = texto.split()
print("4. Lista de palabras:", lista_palabras)

# 5. Sustituye las comas por puntos y comas
texto_sin_comas = texto.replace(",", ";")
print("5. Reemplazar comas:", texto_sin_comas)

# 6. Elimina las 'a' minúsculas
texto_sin_as = texto.replace("a", "")
print("6. Sin 'a' minúsculas:", texto_sin_as)
print()
```

## Ejercicio 3
1. Crea una lista con 3 elementos numéricos
2. Añade un cuarto elemento
3. Calcula la suma de todos
4. Elimina el segundo elemento de la lista
5. Añade otro elemento en la posicion 3 de la lista
6. Crea otra lista con 4 elementos y concaténala a la que ya tenías.
7. Ordena la lista de menor a mayor.
8. BONUS: Ordena la lista de mayor a menor.  
(*PISTA: Para hacer 8, tendrás que buscar en la información del método...*)  
  
  
NOTA: Pinta por pantalla el resultado de cada paso.

```python
# 1. Crea una lista con 3 elementos numéricos
mi_lista = [2, 4, 6]
print("1. Lista inicial:", mi_lista)

# 2. Añade un cuarto elemento
mi_lista.append(8)
print("2. Añadir elemento:", mi_lista)

# 3. Calcula la suma de todos
suma_lista = sum(mi_lista)
print("3. Suma de elementos:", suma_lista)

# 4. Elimina el segundo elemento de la lista (índice 1)
eliminado = mi_lista.pop(1)  # Elimina el 4
print(f"4. Eliminar 2º elemento ({eliminado}):", mi_lista)

# 5. Añade otro elemento en la posición 3 de la lista
mi_lista.insert(2, 5)  # Inserta 5 en la posición 2 (tercer elemento)
print("5. Insertar en posición 3:", mi_lista)

# 6. Crea otra lista con 4 elementos y concaténala
otra_lista = [10, 12, 14, 16]
mi_lista.extend(otra_lista)
print("6. Concatenar listas:", mi_lista)

# 7. Ordena la lista de menor a mayor
mi_lista.sort()
print("7. Ordenar menor a mayor:", mi_lista)

# 8. BONUS: Ordena la lista de mayor a menor
mi_lista.sort(reverse=True)
print("8. BONUS - Ordenar mayor a menor:", mi_lista)
```
